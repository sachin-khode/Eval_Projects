```python
df_income = df['income']
```

```python
df_income.head()
```

```python
df = pd.concat([df, pd.get_dummies(df['income'])], axis=1)
```

```python
df.head()
```

```python
df['income_binary'] = np.nan
df.loc[df['income'] == '>50K', 'income_binary'] = 1
df['income_binary'].fillna(0, inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.hist(figsize=(12,8))
```

```python
from sklearn.preprocessing import LabelEncoder

categorical_features = list(df.select_dtypes(include=['object']).columns)
print("cat feat is here---------->",categorical_features)
label_encoder_feat = {}
for i, feature in enumerate(categorical_features):
#     print("i---->",list(i)
#     print("feature---->", list(feature))
    label_encoder_feat[feature] = LabelEncoder()
    df[feature] = label_encoder_feat[feature].fit_transform(df[feature])

df.head()
```

```python
df_x = df.drop(['income','<=50K','income_binary','>50K'],axis=1)
df_x.head()
```

```python
df_x.shape
```

```python
df_y = df['income']
df_y.head()
```

```python
df_y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_x, df_y,test_size=0.2,random_state= 42)
```

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
lg.fit(X_train,y_train)
```

```python
pred = lg.predict(X_test)
print(pred)
```

```python
lg.score(X_train,y_train)
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
accuracy_score(pred,y_test)
```
