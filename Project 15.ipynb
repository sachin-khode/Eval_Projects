```python
df_x = df.drop(['mpg','car name'],axis=1)
df_x.head(50)
```

```python
df_y = df['mpg']
df_y.head()
```

```python
df_y.shape
```

```python
df_y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error as mse
```

```python
lr=LinearRegression()
rs = []
for i in range(0,100):
        x_train,x_test,y_train,y_test=train_test_split(df_x,df_y,random_state=i,test_size=0.15)
        lr.fit(x_train,y_train)
        y_pred=lr.predict(x_test)
        rscore=r2_score(y_test,y_pred)
        print('random state at',i, 'r2score is:' ,rscore)
        rs.append(rscore)
# print(max(rs))
print('random state for max rscore',rs.index(max(rs)))
    
```

```python
x_train,x_test,y_train,y_test=train_test_split(df_x,df_y,random_state=60,test_size=0.15)
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
rscore=r2_score(y_test,y_pred)
print('r2 score is---->',rscore)
```

```python
rs=[]
for i in range(0,100):
    from sklearn.ensemble import RandomForestRegressor as RF
    RF=RF(random_state=i)
    RF.fit(x_train,y_train)
    y_pred=RF.predict(x_test)
    rscore=r2_score(y_test,y_pred)
    rs.append(rscore)
    print('rscore at',i,'is:',rscore)
print('random state for max rscore',rs.index(max(rs)))
    
```
